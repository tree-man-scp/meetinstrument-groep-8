```python
from mijngeluid import fft_partitioned
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

filenum = 1
fs = 48000
```

```python
#Vul bij str() 1 in, en kijk naar de twee plotjes. en heraal dit voor de hoeveelheid meetingen u heeft gedaan
soundarray = np.load("array_final" + str('14') + ".npy")
nagalmarray = np.load("arr3" + str("14") + ".npy")
```

```python
fsoundarray = np.concatenate((soundarray, nagalmarray))
df = fft_partitioned(fsoundarray[:,0], fs, t_tot=fsoundarray.shape[0]/fs, t_bin=1e-2)
```

```python
#in deze plot staat in deze tabel staat de frequentie met de amplitude erbij
plt.figure(dpi=100)
#ax = plt.axes()
#ax.set_facecolor("black") # voor het gebruikersgemak
plt.xlim([10500,10650])
plt.scatter(df.f, df.t, c=df.I, cmap='magma', marker='.' ,s=1)
plt.colorbar(label="Amplitude [a.u.]", norm=True)
plt.xlabel("Frequentie $f$ [Hz]")
plt.ylabel("Tijd $t$ [s]")
```

```python
t=df.t
I = 10*np.log10((df.I/(10*10**-12)))
```
